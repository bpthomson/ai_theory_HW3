```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Load your training CSV files.
red_train = pd.read_csv("data/winequality-red_train.csv")
white_train = pd.read_csv("data/winequality-white_train.csv")

def plot_column_distributions(df, title):
    """Plot histograms with KDE for every column in the DataFrame."""
    n_features = len(df.columns)
    # We use 3 columns per row; calculate the required number of rows.
    n_cols = 3
    n_rows = math.ceil(n_features / n_cols)
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))
    axes = axes.flatten()  # Flatten grid for easier iteration
    
    for i, col in enumerate(df.columns):
        sns.histplot(df[col], kde=True, ax=axes[i], bins=30)
        axes[i].set_title(f'Distribution of {col}')

    
    fig.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# Visualize distributions for the red wine training data
plot_column_distributions(red_train, "Red Wine Training Data Distributions")

# Visualize distributions for the white wine training data
plot_column_distributions(white_train, "White Wine Training Data Distributions")
```
